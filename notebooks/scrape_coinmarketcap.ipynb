```python
import pandas as pd
import requests
import re
from bs4 import BeautifulSoup
import time
# Create date range for historical snapshots
Date = pd.date_range(start='20130428', end='20210516', freq='7D').strftime('%Y%m%d')

# Retrieve market cap value in dollars

for date in Date:
    if date not in all_data.keys():
        print(date)
        retry = True
        while retry:
            # Retrieve historical snapshot data from date
            try:
                page = requests.get('https://coinmarketcap.com/historical/'+ date +"/")
            except:
                continue
            soup = BeautifulSoup(page.content, 'html.parser')
            try:
                d = return_crypto_data(soup)
                if d:
                    all_data[date] = d
                    retry = False
                else:
                    break
            except Exception as e:
                print(e)
                time.sleep(60)

        time.sleep(5)
    
```

```python
def extract_table_data_from_soup(soup):

    ds = []
    for table_row in soup.select("table tbody tr"):
        
        market_cap_text = table_row.select("td p")[0].text
        d = {"market_cap": market_cap_text}
        for t, cell in zip(["rank","name","symbol", "price","supply","%1h","%24","%7d"],  table_row.select("td div")):
            if t != "price":
                d[t] = cell.text
        ds.append(d)
    return ds
                           
print(extract_table_data_from_soup(soup))
    
    
```
