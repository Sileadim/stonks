```python
from efficient_portfolio.efficient_portfolio import plot_returns, get_actual_returns, load_data, get_max_sharp_allocation


prices, market_cap = load_data()

    
```

```python
import copy

for start, end in [(-52,None),(-104,-52),(-156,-104),(-208,-156), (-264,-208)]:
        df = copy.deepcopy(prices)
        biggest = market_cap.iloc[start].dropna().sort_values(ascending=False)[:40]
        coins = list(biggest.index)
        if coins:
            for c in df.columns:
                if c not in coins:
                    del df[c]
        evaluation_df = df.iloc[start:end]
        
        print("start eval: ", evaluation_df.index[0])
        print("end eval: ", evaluation_df.index[-1])
        
        opti_end = end-52 if end else -52
        optimization_df = df.iloc[start-52:opti_end]
        
        print("start opti: ", optimization_df.index[0])
        print("end opti :", optimization_df.index[-1])
        allocation, sharp = get_max_sharp_allocation(optimization_df, -52, None )
        print("Sharp: ",sharp)
        print("Max sharp previous year: ", allocation)
        max_sharp_returns = get_actual_returns(copy.deepcopy(evaluation_df),allocation )
        plot_returns(max_sharp_returns,"Max sharp")
        market_cap_allocation = (biggest / biggest.sum()).to_frame()
        print("market cap", market_cap_allocation)
        market_cap_returns = get_actual_returns(copy.deepcopy(evaluation_df), market_cap_allocation   )
        plot_returns(market_cap_returns,"Market cap")   
```
